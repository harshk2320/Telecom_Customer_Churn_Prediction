In this notebook we will basically upload the data into MongoDB so that we can extract the data from it later on for our pipeline.

```python
import pandas as pd
import pymongo
from dotenv import load_dotenv
import os
```

```python
df = pd.read_csv("Telco_Customer_Churn.csv")
# df.head(2)
```

```python
# df should be converted into dict before we push it to MongoDB

data = df.to_dict(orient= "records")
```

```python
# Load the .env file
load_dotenv()
```

```python
DB_NAME = os.getenv("DB_NAME")
COLLECTION_NAME = os.getenv("COLLECTION_NAME")
CONNECTION_URL = os.getenv("CONNECTION_URL")
```

```python
DB_NAME, COLLECTION_NAME, CONNECTION_URL
```

```python
client = pymongo.MongoClient(CONNECTION_URL)    
# You are creating a connection object to MongoDB.
# This line basically tell python to connect to connect to my MongoDB cluster using this URL.

data_base = client[DB_NAME]
# Inside your MongoDB cluster, you are selecting the cluster you want to use.
# Basically it tell MongoDB to Open (or create if not exists) the database name <DB_NAME>.env

collection = data_base[COLLECTION_NAME]
# Inside the database, you are selecting a collection (equivalent to a table in SQL).
# If it does not exists, MongoDB will create it automatically when inserting the data.
```

```python
# Uploading data to MongoDB
rec = collection.insert_many(data)  
# insert_many() is a MongoDB funtion which takes list of dictionaries and inserts them into MongoDB Collection.
```
